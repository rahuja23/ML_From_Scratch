```python
import pandas as pd
import math
import numpy as np
import plotly.express as px
import pickle
from src.model import LinearRegression
from src.evaluation import RegressionMetrics
from src.preprocessing import Standardize_data
```

```python
# Loading training and testing data
train_data = pd.read_csv("/Users/rachitahuja/Documents/ML_Bsics/Linear_regression/dataset/train.csv")
test_data = pd.read_csv("/Users/rachitahuja/Documents/ML_Bsics/Linear_regression/dataset/test.csv")

# Remove missing values from both the datasets 
train_data = train_data.dropna()
test_data = test_data.dropna()
```

```python
X_train = train_data["x"].values
y_train = train_data["y"].values

X_test = test_data["x"].values
y_test = test_data["y"].values
```

```python
X_train = Standardize_data(X_train)
X_test = Standardize_data(X_test)
```

```python
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)
```

```python
lr = LinearRegression(0.01)
lr.fit(X_train, y_train, 1000)
```

```python
lr.save_model('model.pkl')
```

```python
model = LinearRegression.load_model("model.pkl")
```

```python
y_pred = model.predict(X_test)
mse_value = RegressionMetrics.mean_squared_error(y_test, y_pred)
rmse_value = RegressionMetrics.root_mean_squared_error(y_test, y_pred)
r_squared_value = RegressionMetrics.r_squared(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse_value}")
print(f"Root Mean Squared Error (RMSE): {rmse_value}")
print(f"R-squared (Coefficient of Determination): {r_squared_value}")
```
